```python
import pandas as pd
import streamlit as st
import os

st.title("Debt & Payment Dashboard")

file_name = "transactions.csv"

# Create CSV if it does not exist
if not os.path.exists(file_name):
    df_init = pd.DataFrame(columns=["date","type","amount"])
    df_init.to_csv(file_name, index=False)

# Load data
df = pd.read_csv(file_name)

if not df.empty:
    df["date"] = pd.to_datetime(df["date"])
    df = df.sort_values("date")
    df["balance"] = df["amount"].cumsum()

# --- Add transaction form ---
st.subheader("Add Transaction")

with st.form("transaction_form"):
    
    date = st.date_input("Date")
    
    type_option = st.selectbox(
        "Transaction Type",
        ["debt","payment"]
    )
    
    amount = st.number_input("Amount", step=1000)

    submitted = st.form_submit_button("Add")

    if submitted:
        if type_option == "debt":
            amount = -abs(amount)
        else:
            amount = abs(amount)

        new_row = pd.DataFrame(
            [[date, type_option, amount]],
            columns=["date","type","amount"]
        )

        new_row.to_csv(file_name, mode="a", header=False, index=False)

        st.success("Transaction Added")
        st.rerun()

# Reload updated data
df = pd.read_csv(file_name)

if not df.empty:
    df["date"] = pd.to_datetime(df["date"])
    df = df.sort_values("date")
    df["balance"] = df["amount"].cumsum()

    # Metrics
    total_debt = df[df["type"]=="debt"]["amount"].sum()
    total_payments = df[df["type"]=="payment"]["amount"].sum()
    current_balance = df["balance"].iloc[-1]

    col1,col2,col3 = st.columns(3)

    col1.metric("Total Debt", f"¥{abs(total_debt):,}")
    col2.metric("Total Payments", f"¥{total_payments:,}")
    col3.metric("Current Balance", f"¥{current_balance:,}")

    st.divider()

    st.subheader("Transactions")
    st.dataframe(df)

    st.subheader("Balance Over Time")
    st.line_chart(df.set_index("date")["balance"])

    st.subheader("Totals by Type")
    st.bar_chart(df.groupby("type")["amount"].sum())
```
